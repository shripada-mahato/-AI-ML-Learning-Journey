find the difference between largest & smallest number

```python
arr = [12, 5, 27, 3, 18]

largest_no = 0
for i in arr:
    if i > largest_no:
        largest_no = i

smallest_no = arr[0]
for i in arr:
    if i < smallest_no:
        smallest_no = i

difference = largest_no - smallest_no

print("largest no is ",largest_no)
print("smallest no is ",smallest_no)
print("difference is ",difference)
```
