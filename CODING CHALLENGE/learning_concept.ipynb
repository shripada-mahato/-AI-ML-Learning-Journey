Dictionary Mini Challenge 6 (Frequency Counter)

```python
arr = [1, 2, 2, 3, 3, 3]
freq = {}

for i in arr:
    if i not in freq:
        freq[i] = 1
print(freq)
```

Challenge 27: Find the Longest Word

```python
text = "machine learning is awesome in python programming"

word = text.split()
longest = word[0]

for i in word:
    if len(i) > len(longest):
        longest = i
print(word)
print(longest)

```
